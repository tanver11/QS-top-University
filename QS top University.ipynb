institutions_by_countries = qs.groupby('Location') \

```python
institutions_by_countries
```

```python
x = institutions_by_countries["Location"]
y = institutions_by_countries["number_of_institutions"]
plt.title = 'Institution by countries'
plt.xticks(rotation = 70)
plt.xlabel('Country')
plt.ylabel('Institute Number')
plt.bar(x,y)
```
